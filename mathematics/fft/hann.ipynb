```python
import numpy as np
import scipy.signal as sl
import scipy.integrate as si
import pandas as pd
```

```python
import seaborn as sns
sns.set_style('whitegrid')
```

```python
n = 1000

x = np.linspace(0, 1, n)
f = sl.windows.hann(2 * n)[n:]

F = si.cumtrapz(f, x=x, initial=0)

s = pd.DataFrame({'f': f, 'F': F}, index=x)
```

```python
s.plot()
```
