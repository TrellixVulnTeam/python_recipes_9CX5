```python
import sobol as sobol_sequence
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
# %config InlineBackend.figure_format = 'svg'
```

```python
nsamples = 250
```

```python
qseq = sobol_sequence.sample(nsamples, 2)
qseq[:10, 0]
```

```python
pseq = np.random.rand(nsamples, 2)
pseq[:10, 0]
```

```python
pqseq = np.zeros_like(qseq)
pqseq = (qseq + np.random.rand(1)) % 1
```

```python
fig, axs = plt.subplots(ncols=3, sharey=True, 
                        subplot_kw={'aspect': 1}, 
                        figsize=(8, 8))
axs[0].plot(*qseq.T, '*')
axs[1].plot(*pseq.T, '*')
axs[2].plot(*pqseq.T, '*')
plt.show()
```
