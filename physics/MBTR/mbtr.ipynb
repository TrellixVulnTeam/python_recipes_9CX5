```python
import qmmlpack as qmml
from ase.io import read
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
trm = read('trm.in', -1, 'aims')
ltrm = read('ltrm.in', -1, 'aims')
```

```python
# ?qmml.many_body_tensor
```

```python
z=[trm.numbers]
r_trm=[trm.positions]
r_ltrm=[ltrm.positions]
b_trm=[trm.cell]
b_ltrm=[ltrm.cell]
kgwdcea=[2, '1/distance', 'identity', ('normal', [0.0025]), 'identity', 'noreversals', 'noreversals']
d=[0.2, .001, 500]

qm_trm = qmml.many_body_tensor(z=z, r=r_trm, basis=b_trm, d=d, kgwdcea=kgwdcea)
qm_ltrm = qmml.many_body_tensor(z=z, r=r_ltrm, basis=b_ltrm, d=d, kgwdcea=kgwdcea)
```

```python
x = np.arange(d[0], d[0] + d[1]*d[2], d[1])
```

```python
fig, axs = plt.subplots(2, sharex=True)
axs[0].plot(1/x, qm_trm[0][0])
axs[1].plot(1/x, qm_ltrm[0][0])
```
